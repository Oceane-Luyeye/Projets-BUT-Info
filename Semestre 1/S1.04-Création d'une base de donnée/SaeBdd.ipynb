Pour finir , nous avons créer la table TITANIC dans le bus de pourvoir reliée la table passagers et embarquement. Cela nous a aider à créer les clés étrangère.


```python
%%sql

CREATE TABLE TITANIC (PassagerID int PRIMARY KEY,
                      Survived int ,
                      Pclass int , 
                      Name varchar(150),
                      Sex varchar(50),
                      Age float ,SibSp int, 
                      Parch int , 
                      Ticket varchar (150), 
                      Fare float, 
                      Cabin varchar(150), 
                      Embarked varchar (150));
```

```python
%%sql 
COPY TITANIC FROM '/Users/oceane/Downloads/titanic_train.csv' DELIMITER','CSV HEADER;
```

```python
%%sql

CREATE TABLE passagers (PassagerID integer PRIMARY KEY,
                       Survived integer,
                       Pclass integer,
                       Name varchar(150),
                       Sex varchar(50),
                       Age float,
                       SibSp integer,
                       Parch integer, 
                       foreign key (PassagerID) references TITANIC (PassagerID));
```

requete 5

```python
%%sql

Create view survivant as 
select  survived , pclass , COUNT(*)
from passagers 
where survived ='0' or survived ='1' 
GROUP BY  survived , pclass;
```

```python
%%sql

select* from survivant;
```
